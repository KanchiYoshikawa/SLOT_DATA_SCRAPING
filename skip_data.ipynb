```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import Select
from xlwt import Workbook
import time
import urllib.request
import os
import datetime
import openpyxl
import glob



def Output_excel(sheet, count=2, youbi=1):
    #台数の取得
    element = driver.find_elements_by_xpath("/html/body/div[3]/div[2]/article[2]/section[1]/div[3]/div[2]/table/tbody/tr")
    machine_num = len(element)
    
    time.sleep(1)

    try:
        #日付の変更
        element = driver.find_element_by_name("target_date")
        select = Select(element)
        select.select_by_index(youbi)

        time.sleep(1)
        element = driver.find_element_by_name("target_date")
        select = Select(element)
        dt_now = select.first_selected_option.text

        #データの収集
        for j in range(1, machine_num + 1, 1):

            #機種名の記入
            element = driver.find_element_by_xpath("/html/body/div[3]/div[2]/section/article/div[1]/div/span/p")
            sheet["A" + str(count + j)] = str(element.text)

            #台番号
            element = driver.find_element_by_xpath("/html/body/div[3]/div[2]/article[2]/section[1]/div[3]/div[2]/table/tbody/tr["+ str(j) + "]/td[1]/a")
            sheet["B" + str(count + j)] = int(element.text)

            #総回転数 /html/body/div[3]/div[2]/article[2]/section[1]/div[3]/div[3]/table/tbody/tr[2]/td[7]
            element = driver.find_element_by_xpath("/html/body/div[3]/div[2]/article[2]/section[1]/div[3]/div[2]/table/tbody/tr["+ str(j) + "]/td[6]")
            try:
                sheet["C" + str(count + j)] = int(element.get_attribute("textContent"))
            except:
                sheet["C" + str(count + j)] = "--"

            #BB回数
            element = driver.find_element_by_xpath("/html/body/div[3]/div[2]/article[2]/section[1]/div[3]/div[2]/table/tbody/tr["+ str(j) + "]/td[2]")
            try:
                sheet["D" + str(count + j)] = int(element.get_attribute("textContent"))
            except:
                sheet["D" + str(count + j)] = "--"

            #RB回数
            element = driver.find_element_by_xpath("/html/body/div[3]/div[2]/article[2]/section[1]/div[3]/div[2]/table/tbody/tr["+ str(j) + "]/td[3]")
            try:
                sheet["E" + str(count + j)] = int(element.get_attribute("textContent"))
            except:
                sheet["E" + str(count + j)] = "--"

            #ART回数
            element = driver.find_element_by_xpath("/html/body/div[3]/div[2]/article[2]/section[1]/div[3]/div[2]/table/tbody/tr["+ str(j) + "]/td[4]")
            try :
                sheet["F" + str(count + j)] = int(element.get_attribute("textContent"))
            except :
                sheet["F" + str(count + j)] = "--"

            #BB確率
            element = driver.find_element_by_xpath("/html/body/div[3]/div[2]/article[2]/section[1]/div[3]/div[2]/table/tbody/tr["+ str(j) + "]/td[7]")
            try :
                rate = element.get_attribute("textContent")
                sheet["G" + str(count + j)] = float(rate[2:])
            except :
                sheet["G" + str(count + j)] = "--"

            #RB確率
            element = driver.find_element_by_xpath("/html/body/div[3]/div[2]/article[2]/section[1]/div[3]/div[2]/table/tbody/tr["+ str(j) + "]/td[8]")
            try :
                rate = element.get_attribute("textContent")
                sheet["H" + str(count + j)] = float(rate[2:])
            except :
                sheet["H" + str(count + j)] = "--"

            #合成確率
            element = driver.find_element_by_xpath("/html/body/div[3]/div[2]/article[2]/section[1]/div[3]/div[2]/table/tbody/tr["+ str(j) + "]/td[9]")
            try :
                rate = element.get_attribute("textContent")
                sheet["I" + str(count + j)] = float(rate[2:])
            except :
                sheet["I" + str(count + j)] = "--"

            #差玉
            element = driver.find_element_by_xpath("/html/body/div[3]/div[2]/article[2]/section[1]/div[3]/div[2]/table/tbody/tr["+ str(j) + "]/td[12]")
            try :
                sheet["J" + str(count + j)] = int(element.get_attribute("textContent"))
            except :
                sheet["J" + str(count + j)] = "--"
    except:
        return None

    return dt_now



URL = "https://m.skipskip.jp/shinyokob1f/"

#URLのページを開く
options = Options()
options.add_argument('--headless')
driver = webdriver.Chrome('.\chromedriver.exe')
driver.implicitly_wait(40)
driver.get(URL)

#Windowのサイズを最大化する
driver.maximize_window()


for youbi in range(1, 8, 1):
    count = 2
    #データ格納Excelの作成
    wb_all = openpyxl.Workbook()
    sheet_all = wb_all['Sheet']

    sheet_all["A2"] = "機種名"
    sheet_all["B2"] = "台番号"
    sheet_all["C2"] = "総回転数"
    sheet_all["D2"] = "BB回数"
    sheet_all["E2"] = "RB回数"
    sheet_all["F2"] = "ART回数"
    sheet_all["G2"] = "BB確率"
    sheet_all["H2"] = "RB確率"
    sheet_all["I2"] = "合成確率"
    sheet_all["J2"] = "差玉"
    
    #46円スロ選択
    element = driver.find_element_by_xpath("/html/body/div[1]/div/article/div[3]/div/div[1]/a/div")
    actions = ActionChains(driver)
    actions.move_to_element(element)
    actions.perform()
    element.click()
    
    element = driver.find_elements_by_xpath("/html/body/div[1]/div[2]/article/section[1]/div[2]/div/table/tbody/tr")
    hole_number = len(element)
    time.sleep(1)
    
    element = driver.find_element_by_xpath("/html/body/div[1]/div[1]/nav[2]/div/div/h1/a")
    actions = ActionChains(driver)
    actions.move_to_element(element)
    actions.perform()
    element.click()

    for i in range(1, hole_number+1, 1):
        #46円スロ選択
        element = driver.find_element_by_xpath("/html/body/div[1]/div/article/div[3]/div/div[1]/a/div")
        actions = ActionChains(driver)
        actions.move_to_element(element)
        actions.perform()
        element.click()

        #1機種目の選択
        time.sleep(1)
        element = driver.find_element_by_xpath("/html/body/div[1]/div[2]/article/section[1]/div[2]/div/table/tbody/tr[" + str(i) + "]/td[1]/span/a")
        actions = ActionChains(driver)
        actions.move_to_element(element)
        actions.perform()
        element.click()

        #全データのexcelのデータ収集
        #データを取得しexcelへの記入
        element = driver.find_elements_by_xpath("/html/body/div[3]/div[2]/article[2]/section[1]/div[3]/div[2]/table/tbody/tr")
        machine_num = len(element)

        dt_now = Output_excel(sheet_all, count, youbi)
        
        if(dt_now != None):
            hole_dir = "C:/Users/kanch/Documents/Crawler/program/skip"
            excel_name_all = hole_dir + "/" + dt_now.replace('/', '-') + ".xlsx"

            #excelファイルの保存
            wb_all.save(excel_name_all)
            count += machine_num
        
        element = driver.find_element_by_xpath("/html/body/div[3]/div[1]/nav[2]/div/div/h1/a")
        actions = ActionChains(driver)
        actions.move_to_element(element)
        actions.perform()
        element.click()
```
